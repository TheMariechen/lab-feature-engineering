There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
spaceship.dropna(inplace=True)
```

```python
spaceship.isna().any()
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
spaceship["Cabin"]
```

```python
spaceship["Cabin"] = spaceship["Cabin"].apply(lambda x: x.split("/")[0])
spaceship["Cabin"]
```

- Drop PassengerId and Name

```python
spaceship.drop(['Name','PassengerId'],axis=1,inplace=True)
```

```python
spaceship.head()
```

- For non-numerical columns, do dummies.

```python
spaceship.dtypes
```

```python
# First identifying non-numerical columns: Home Planet, Destination 
# Boolean only need to be changed if object type not correctly recognised! typed as object in the case of: Cryosleep, VIP
```

```python
spaceship["HomePlanet"].value_counts()
```

```python
spaceship["Destination"].value_counts()
```

```python
spaceship = pd.get_dummies(spaceship, columns = ['Destination', 'HomePlanet', 'Cabin'])
spaceship
```

```python
# option with drop first - meaning that the first of the created dummy columns is removed and treated as baseline! because lost in table but 
spaceship = pd.get_dummies(spaceship, columns = ['Destination', 'HomePlanet', 'Cabin'], drop_first = True) 
spaceship
```

```python
spaceship["CryoSleep"] = spaceship["CryoSleep"].apply(lambda x: 1 if x == True else 0)
```

```python
spaceship["CryoSleep"].unique()
```

```python
spaceship["VIP"] = spaceship["VIP"].apply(lambda x: 1 if x == True else 0)
```

```python
spaceship["VIP"].unique()
```

```python
spaceship.dtypes 
```

```python
spaceship
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

**Perform Train Test Split**

```python
features = spaceship.drop(columns = ["Transported"]) # dropping target column to keep evreything else!
target = spaceship["Transported"] # defining target column 
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=0)
```

```python
# Performing normal KNN without normalization first!
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
# Model fit without normalization: 0.7700453857791225
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
# Following steps to first normalise the data before initialising the model
normalizer = MinMaxScaler()
```

```python
normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_train_norm.head()
```

```python
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_test_norm.head()
```

```python
# Initializing
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
# Training
knn.fit(X_train_norm, y_train)
```

- Evaluate your model's performance. Comment it

```python
# Evaluating 
knn.score(X_test_norm, y_test) # this line always gives us accuracy as evaluation metric for classifications (as opposed top R2 in regressions) 
# With normalization = 0.7609682299546142, not big difference
```

2. Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Initialize
tree = DecisionTreeRegressor(max_depth=3)
```

```python
# Training 
tree.fit(X_train, y_train)
```

```python
# Evaluating
X_train
```

```python
y_test
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
pred = tree.predict(X_test)

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test, squared=False))
print("R2 score", tree.score(X_test, y_test))
# Model explains only 31% of the variation - not so well suited for the data
```

```python
# Additional check: which were the most relevant features? 
tree_importance = {feature : importance for feature, importance in zip(X_train_norm.columns, tree.feature_importances_)}
tree_importance # but how exactly to interpret? 
```

```python
# Displaying decision tree visually
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (34,20))
plot_tree(tree,filled = True, rounded=True,feature_names=list(X_train.columns))
plt.show()
```
